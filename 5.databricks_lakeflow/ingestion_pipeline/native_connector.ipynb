#How to create a native pipeline:
1. Create a sales force account(already done by  company)
2. Create a connection (Catalog->create connection)
a. connection name - give some name for connection
b. Connection type - give the connector's name(eg salesforce)
c. Login into Salesforce to establish the connection
3. Create a ingestion pipeline:Data Engineering-> data ingestion -> select salesforce
a. select the connection created previously
b. Ingestion setup: pipeline name, event_log(catalog and schema)
c. select the source tables to ingest 
d. select the target schema and catalog to create the tables
e. Create the schedule for ingestion
4. Once the pipeline is created we can check the 'jobs and pipelines' and select the job to view its run and its details
