```scala
%python
dbutils.widgets.text("catalog","new_catalog")
catalog=dbutils.widgets.get("catalog")
dbutils.widgets.text("schema","default")
schema=dbutils.widgets.get("schema")
```

###Creating catalog and schema

```scala
%python
print(f"Creating '{catalog}' catalog")
spark.sql(f"CREATE CATALOG IF NOT EXISTS {catalog}")
print(f"Creating '{schema}' schema")
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {schema}")
```

##Creating silver,bronze,gold and datasource volumes

```scala
%python
print(f"Creating gold volume in {catalog}.{schema}")
spark.sql(f"CREATE VOLUME IF NOT EXISTS {catalog}.{schema}.gold")
print(f"Creating silver volume in {catalog}.{schema}")
spark.sql(f"CREATE VOLUME IF NOT EXISTS {catalog}.{schema}.silver")
print(f"Creating bronze volume in {catalog}.{schema}")
spark.sql(f"CREATE VOLUME IF NOT EXISTS {catalog}.{schema}.bronze")
print(f"Creating sourcedata volume in {catalog}.{schema}")
spark.sql(f"CREATE VOLUME IF NOT EXISTS {catalog}.{schema}.sourcedata")
```

##Defining variables for paths and databases

```scala
%python
gold_path=f"/Volumes/{catalog}/{schema}/gold"
print(f"The gold volume is located at {gold_path}")
silver_path=f"/Volumes/{catalog}/{schema}/silver"
print(f"The silver volume is located at {silver_path}")
bronze_path=f"/Volumes/{catalog}/{schema}/bronze"
print(f"The bronze volume is located at {bronze_path}")
sourcedata_path=f"/Volumes/{catalog}/{schema}/sourcedata"
print(f"The sourcedata volume is located at {sourcedata_path}")
silver_db=f"{catalog}.{schema}"
print(f"The silver database is {silver_db}")
gold_db=f"{catalog}.{schema}"
print(f"The gold database is {gold_db}")
```

##Returning variables to calling nb

```scala
%python
import json
dbutils.notebook.exit(
    json.dumps(
        {
            "catalog":catalog,
            "schema":schema,
            "gold_path": gold_path,
            "silver_path": silver_path,
            "bronze_path": bronze_path,
            "sourcedata_path": sourcedata_path,
            "silver_db": silver_db,
            "gold_db": gold_db,
        })
)
```
