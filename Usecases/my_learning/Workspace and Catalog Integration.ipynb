###Workspace and Catalog in Databricks

####Workspace

A workspace is your working environment in Databricks.

Inside a workspace, we have:
- Notebooks (Python, SQL, Scala, R)
- Jobs & workflows
- Clusters / SQL Warehouses
- Dashboards
- Repo integrations (Git)
- Users & permissions (workspace-level)

Key points
- Multiple teams usually have multiple workspaces
- Workspaces are isolated for compute & UI
- Data is not stored in the workspace itself

####Workspace and Catalog integration
**Access to Workspace**
- You are added to a workspace
- This lets you:
Create notebooks
Run jobs
Attach clusters

üëâ This does NOT give data access by itself.

**Workspace Attached to Catalog (Metastore)**
- The workspace is attached to a Unity Catalog metastore
- This makes catalogs visible in the workspace
- Example: SHOW CATALOGS;

üëâ Workspace can see catalogs, but cannot use them without permission

**User Permission on Catalog Objects (Most Important)**

Unity Catalog permissions are granted to:
- Users
- Groups
- Service principals

Example:
GRANT USE CATALOG ON CATALOG sales_catalog TO `data_engineers`;
GRANT USE SCHEMA ON SCHEMA sales_catalog.gold TO `data_engineers`;
GRANT SELECT ON TABLE sales_catalog.gold.daily_revenue TO `data_engineers`;
üìå Only then can notebooks access the data.

**So What Happens When You Run a Notebook?**
When you run a notebook:
- Notebook runs in a workspace
- Cluster checks your identity
- Unity Catalog checks your permissions
- If allowed ‚Üí access granted
- If not ‚Üí Permission denied
