```scala
%python
from pyspark.sql.functions import col,to_date,try_to_date,try_to_timestamp,coalesce
df1=spark.read.csv("/Volumes/myowndatasets/malformed_data/malformed_csv/malformed_csv1.csv",header=True,inferSchema=True,comment='#',quote='"',ignoreLeadingWhiteSpace=True,ignoreTrailingWhiteSpace=True,mode="PERMISSIVE",columnNameOfCorruptRecord="corrupt_record",multiLine=True)

df2= df1.withColumn("join_date_formatted", try_to_date(col("join_date"), "yyyy-MM-dd"))
df3 = df1.withColumn(
    "last_login_formatted",
    coalesce(
        try_to_timestamp(col("last_login"), "yyyy-MM-dd HH:mm:ss"),
        try_to_timestamp(col("last_login"), "yyyy-MM-dd HH:mm:ssXXX"),
        try_to_timestamp(col("last_login"), "yyyy-MM-dd hh:mm a"),
        try_to_timestamp(col("last_login"), "dd-MM-yyyy HH:mm"),
        try_to_timestamp(col("last_login"), "dd/MM/yy HH:mm"),
        try_to_timestamp(col("last_login"), "yyyy-MM-dd H:m")
    )
)

display(df3)
```
