```scala
%python
df1=[(1,"Joe"),(2,"Frank"),(3,"peter")]
schema=["cust_id","name"]
tbl1=spark.createDataFrame(data=df1,schema=schema)
```

```scala
%python
df2=[{"txns_id":1,"cust_id":1,"amount":50},{"txns_id":2,"cust_id":2,"amount":60},{"txns_id":3,"cust_id":2,"amount":70},{"txns_id":4,"cust_id":1,"amount":80},{"txns_id":5,"cust_id":4,"amount":100}]
schema2=["txns_id","cust_id","amount"]
tbl2=spark.createDataFrame(df2)
tbl2=tbl2.select("txns_id","cust_id","amount")
display(tbl1)
display(tbl2)
```

```scala
%python
display(tbl1.join(tbl2,how="inner",on="cust_id"))
```

```scala
%python
display(tbl1.join(tbl2,how="full_outer",on="cust_id"))
```
