##3. Spark Directory Read Use Cases
1. Read all tower logs using:

Recursive lookup

2. Demonstrate these 3 reads separately:
Using pathGlobFilter
Using list of paths in spark.read.csv([path1, path2])

3. Compare the outputs and understand when each should be used.

```python
df1=spark.read.csv("dbfs:///Volumes/telecom_catalog_assign/landing_zone/landing_vol/tower/*",pathGlobFilter="*.csv",recursiveFileLookup="True",header=True,inferSchema=True,sep='|')
display(df1)
#recursive file loop-> looks into all files and subfolders inside the given folder and takes all files inside irrespective of file type and file  name (for this all the files inside the folder has to be in same format and numbers of columns and data types should match)
#pathGlobFilter-> filters the files based on the given pattern
#inferSchema-> automatically infers the schema of the csv file

```

```python
#PathGlobalFilter in nature is nonrecursive- ie. it looks in the given directory for given pattern matching files
df1=spark.read.csv("dbfs:///Volumes/telecom_catalog_assign/landing_zone/landing_vol/tower/*",pathGlobFilter="*.csv",header=True,inferSchema=True,sep='|')
display(df1)

#PathGlobalFilter with recursive file look up becomes recursive- ie. it looks in the given directory for given pattern matching files and also the all the subdirectories 
df1=spark.read.csv("dbfs:///Volumes/telecom_catalog_assign/landing_zone/landing_vol/tower/*",pathGlobFilter="*.csv",recursiveFileLookup="True",header=True,inferSchema=True,sep='|')
display(df1)

#Instead of recursive file look up we can use multiple paths directly in path variable in form of list

df1=spark.read.csv(["dbfs:///Volumes/telecom_catalog_assign/landing_zone/landing_vol/tower/region1/*","/Volumes/telecom_catalog_assign/landing_zone/landing_vol/tower/region2/*"],header=True,inferSchema=True,sep='|')
display(df1)
```

**When to use multiple paths using path varibale:**
- ✔️ You want specific folders only
- ✔️ You want manual control
- ✔️ Avoid scanning unnecessary data
**When to use recursive file look up with pathglobalfilter using path varibale:**
- ✔️ Deep folder structure(large number of subfolders)
- ✔️ Automatic scanning of all files 
- ✔️ pathglobalfilter -> when folder has different file types

##15. Do a final exercise of defining one/two liner of... 
1. When to use/benifits csv ->
      -       storing raw data
      -       Occupies more space
      -       doesnt work well with gzip unlike xml or json

2. When to use/benifits json -> 
      -       storing api logs
      -       gives better compression with gzip
      -       stroing unstructred data

3. When to use/benifit orc ->
      -       columar file format, 
      -       works well with zstd(better storage) also snappy(speed) to some extent
      -       hive tables
      -       default compression is snappy

4. When to use/benifit parquet
    -         default compression is snappy
    -         columar file type
    -         faster read and write
    -         comparitively greater storage


5. When to use/benifit delta
    -         default parquet file with snappy compression (databricks default)
    -         columar file type
    -         faster read and write
    -         comparitively greater storage
    -         acid trasanctions
    -         time travel with help of delta logs
6. When to use/benifit xml
    - xml with gzip - maximum compression
    - xmls used in legacy systems

7. When to use/benifit delta tables
    - tables with updates
    - acid tables
    - requires timetravel
